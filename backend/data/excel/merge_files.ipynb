```python
df_2 = pd.read_excel("gym_exercise_dataset_cleaned.xlsx")

df_2.drop(
    columns = [
        "id",
        "popularity_score",
        "exercise_tier",
        "created_at",
        "updated_at",
        "Unnamed: 16",
        "Unnamed: 17",
    ],
    inplace=True
)

df_2.rename(columns={"popularity_score_new (1-3)": "popularity_score", "exercise_tier_new": "tier", "instructions": "preparation", "Execution": "execution"}, inplace=True)

df_2
```

```python
# Method 1: See which rows are duplicated (keeps first occurrence, marks others as True)
duplicated_mask = df_2.duplicated(subset=["name", "equipment"], keep=False)
duplicated_rows = df_2[duplicated_mask].sort_values(by=["name", "equipment"], ascending=False)
print(f"Number of duplicate rows: {len(duplicated_rows)}")
print(f"Number of unique duplicate groups: {len(duplicated_rows.groupby(['name', 'equipment']))}")

duplicated_rows.to_excel("duplicated_rows.xlsx", index=False)

duplicated_rows
```

```python
# Create a priority column for the main_muscles we want to keep
df_2['tokeep'] = df_2['main_muscles'].isin(['["Quadriceps"]', '["Anterior Deltoid"]'])

# Sort by priority (True first) and then by other columns to ensure consistent ordering
df_2 = df_2.sort_values(['tokeep', 'name', 'equipment'], ascending=[False, True, True])

# Drop duplicates keeping the first occurrence (which will be the priority ones)
df_2 = df_2.drop_duplicates(subset=["name", "equipment"], keep='first')

# Drop the priority column as it's no longer needed
df_2.drop('tokeep', axis=1, inplace=True)
```
