```python
import os
import sys
from pathlib import Path

# Add backend directory to Python path
# Notebook is in backend/scripts/, so go up one level to get backend/
backend_dir = Path.cwd()
if backend_dir.name == 'scripts':
    backend_dir = backend_dir.parent
elif 'backend' not in str(backend_dir):
    # If we're not in backend, try to find it
    backend_dir = backend_dir / 'backend' if (backend_dir / 'backend').exists() else backend_dir.parent / 'backend'

if str(backend_dir) not in sys.path:
    sys.path.insert(0, str(backend_dir))

from google import genai
from app.config.model_config import DEFAULT_LLM_MODEL_COMPLEX

# Get API key from environment
API_KEY = os.getenv("LLM_API_KEY")
if not API_KEY:
    raise ValueError("LLM_API_KEY not found in environment variables")

# Initialize Gemini client
client = genai.Client(api_key=API_KEY)

# Get model name from config
model_name = DEFAULT_LLM_MODEL_COMPLEX

# Make a simple call
response = client.models.generate_content(
    model=model_name,
    contents="Hello! Say hello back in one sentence.",
)

# Print the response
print(response.text)
```
